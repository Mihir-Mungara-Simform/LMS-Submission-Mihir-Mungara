```python
# Task 5: Abstraction
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def calculate_area():
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
        
    def calculate_area(self):
        return self.length * self.width
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def calculate_area(self):
        return (22 / 7) * self.radius * self.radius
    
rec_obj = Rectangle(5,6)
print(f'area of rectangle : {rec_obj.calculate_area()}')
print()

cir_obj = Circle(7)
print(f'area of circle : {cir_obj.calculate_area()}')
```

```python
# Task 6: Method Overloading
# version 1
class Vehicle:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.__price = price

    def display_info(self):
        print(f'brand:- {self.brand} \nmodel :- {self.model} \nyear :- {self.year} \nprice :- {self.__price}')

    def get_price(self):
        return self.__price

    def set_price(self, value):
        self.__price = value

    def calculate_discount(self, percentage):
        self.__price -= self.__price * percentage / 100


dis_obj1 = Vehicle('xyz', 'M4', 2020, 30000000.00)

print('Before Discount : ')
dis_obj1.display_info()
print()

print('After Discount : ')
dis_obj1.calculate_discount(50)
dis_obj1.display_info()
```

```python
# Task 6: Method Overloading
# version 2
class Vehicle:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.__price = price

    def display_info(self):
        print(f'brand:- {self.brand} \nmodel :- {self.model} \nyear :- {self.year} \nprice :- {self.__price}')

    def get_price(self):
        return self.__price

    def set_price(self, value):
        self.__price = value

    def calculate_discount(self, percentage, additional_discount = None):
        self.__price -= self.__price * percentage / 100
        if additional_discount is not None:
            self.__price -= additional_discount


dis_obj1 = Vehicle('xyz', 'M4', 2020, 30000000.00)

print('Before Discount : ')
dis_obj1.display_info()
print()

print('After Discount : ')
dis_obj1.calculate_discount(50, 10000000)
dis_obj1.display_info()
```

```python
# Task 7: File Handling with OOP
class Vehicle:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.__price = price

    def display_info(self):
        print(f'brand:- {self.brand} \nmodel :- {self.model} \nyear :- {self.year} \nprice :- {self.__price}')

    def get_price(self):
        return self.__price

    def set_price(self, value):
        self.__price = value

    def calculate_discount(self, percentage, additional_discount = None):
        self.__price -= self.__price * percentage / 100
        if additional_discount is not None:
            self.__price -= additional_discount

    def save_to_file(self):
        filename = f"{self.brand}_{self.model}.txt"

        with open(filename, 'w') as f:
            f.write(f"Brand: {self.brand}\n")
            f.write(f"Model: {self.model}\n")
            f.write(f"Year: {self.year}\n")
            f.write(f"Price: {self.__price}\n")
            f.write("-----------------\n")

obj_fh = Vehicle('xyz', 'M4', 2020, 30000000.00)
obj_fh.display_info()
obj_fh.save_to_file()
```

```python
# Task 8: Real-World Application
class Library:
    def __init__(self,list_of_books):
        self.list_of_books = list_of_books

    def display_all_books(self):
        print(self.list_of_books)

    def borrow_book(self, book_name):
        self.list_of_books.remove(book_name)

    def return_book(self, book_name):
        self.list_of_books.append(book_name)
```

```python
ob = Library(['a', 'b', 'c'])

while True:
    print("\nLibrary Menu")
    print("1. Display all book titles")
    print("2. Borrow a book")
    print("3. Return a book")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        print("\nAvailable Books:")
        ob.display_all_books()

    elif choice == '2':
        book_name = input("Enter book to borrow: ")
        ob.borrow_book(book_name)

        print("\nRemaining books:")
        ob.display_all_books()

    elif choice == '3':
        book_name = input("Enter book to return: ")
        ob.return_book(book_name)

        print("\nUpdated books list:")
        ob.display_all_books()

    elif choice == '4':
        print("Exiting program...")
        break

    else:
        print("Invalid choice. Please try again.")
```
